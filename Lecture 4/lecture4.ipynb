```python
import numpy as np

# linear
phi = np.array([
    [1, 1],
    [2, 1],
    [2, 1],
    [3, 1]
])

X = np.array([
    [1],
    [2],
    [2],
    [3]
])

y = np.array([
    [1],
    [1],
    [2],
    [2]
])

w = np.matmul(np.matmul(np.linalg.inv(np.matmul(np.transpose(phi), phi)), np.transpose(phi)), y)
print(w)
```

```python
# quadratic
X = np.array([
    [1],
    [2],
    [1.5],
    [3]
])

phi = np.array([
    [1, 1, 1],
    [4, 2, 1],
    [2.25, 1.5, 1],
    [9, 3, 1]
])

y = np.array([
    [1],
    [1],
    [0],
    [2]
])

w = np.matmul(np.matmul(np.linalg.inv(np.matmul(np.transpose(phi), phi)), np.transpose(phi)), y)
print(w)
```

```python
# plotting
import matplotlib.pyplot as plt 

plt.scatter(X, y)
plt.xlim(0, 3.5)
plt.ylim(0, 3.5)

para_x = np.linspace(-10, 10, 1000)
para_y = w[0]*para_x**2 + w[1]*para_x + w[2]
plt.plot(para_x, para_y)

plt.show()
```
