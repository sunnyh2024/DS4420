```python
# imports
import autograd.numpy as np
from autograd import grad
```

```python
# initializing the matrices
x_1 = np.array([
    [1],
    [0]
])

x_2 = np.array([
    [2],
    [1]
])

x = np.array([
    [1],
    [2]
])

y = w = np.array([
    [2],
    [0]
])

A = np.array([
    [1, 0],
    [0, 2]
])

one = np.array([
    [1],
    [1]
])
```

```python
# part 1
def f(x):
    return np.dot(np.transpose(x), np.matmul(A, x)) - np.dot(np.transpose(x), y)

def df(x):
    return np.array([2*x[0] - 2, 4*x[1]])

auto = grad(f)

for i in range(10):
    t = np.random.random((2, 1))
    print(f'Input x: \n{t}')
    print(f'Auto: \n{auto(t)}\nTheoretical: \n{df(t)}\n')
```

```python
# part 2
def f(x):
    return np.dot(np.transpose(x), np.matmul(A, one))

def df(x):
    return np.array([
        [1],
        [2]
    ])

auto = grad(f)

for i in range(10):
    t = np.random.random((2, 1))
    print(f'Input x: \n{t}')
    print(f'Auto: \n{auto(t)}\nTheoretical: \n{df(t)}\n')
```

```python
# part 3
def f(x: np.array):
    return np.matmul(np.transpose(x), one)

def df(x):
    return np.array([
        [1],
        [1]
    ])

auto = grad(f)

for i in range(10):
    t = np.random.random((2, 1))
    print(f'Input x: \n{t}')
    print(f'Auto: \n{auto(t)}\nTheoretical: \n{df(t)}\n')
```

```python
# part 4
def f(w):
    return np.dot(np.transpose(w), x_1) + np.dot(np.transpose(w), x_2)

def df(x):
    return np.array([
        [3],
        [1]
    ])

auto = grad(f)

for i in range(10):
    t = np.random.random((2, 1))
    print(f'Input x: \n{t}')
    print(f'Auto: \n{auto(t)}\nTheoretical: \n{df(t)}\n')
```
