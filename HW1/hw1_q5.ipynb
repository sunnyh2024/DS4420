```python
# imports
import autograd.numpy as np
from autograd import grad
```

```python
# linear 
def f(x, y):
    return np.dot(np.transpose(x), y)

def df(x, y):
    return y

auto = grad(f)

for i in range(10):
    x = np.random.random((2, 1))
    y = np.random.random((2, 1))
    print(f'Input x: \n{x}')
    print(f'Input y: \n{y}')
    print(f'Auto: \n{auto(x, y)}\nTheoretical: \n{df(x, y)}\n')
```

```python
# quadratic
def f(x, A):
    return np.dot(np.dot(np.transpose(x), A), x)

def df(x, A):
    return np.dot((A + np.transpose(A)), x)

auto = grad(f)

for i in range(10):
    x = np.random.random((2, 1))
    A = np.random.random((2, 2))
    print(f'Input x: \n{x}')
    print(f'Input A: \n{A}')
    print(f'Auto: \n{auto(x, A)}\nTheoretical: \n{df(x, A)}\n')
```

```python
# trace of quadratic
def f(x, A):
    return np.trace(np.dot(np.dot(np.transpose(x), A), x))

def df(x, A):
    return np.dot((A + np.transpose(A)), x)

auto = grad(f)

for i in range(10):
    x = np.random.random((2, 1))
    A = np.random.random((2, 2))
    print(f'Input x: \n{x}')
    print(f'Input A: \n{A}')
    print(f'Auto: \n{auto(x, A)}\nTheoretical: \n{df(x, A)}\n')
```

```python
# ReLU
def f(w, x):
    return np.maximum(np.dot(np.transpose(w), x), 0)

def df(w, x):
    dw = 1 if np.dot(np.transpose(w), x) > 0 else 0
    return dw * x

auto = grad(f)

for i in range(10):
    w = np.random.randint(-10, 10, (2, 1))
    x = np.random.randint(-10, 10, (2, 1))
    print(f'Input w: \n{w}')
    print(f'Input x: \n{x}')
    print(f'Auto: \n{auto(w, x)}\nTheoretical: \n{df(w, x)}\n')
```

```python
# Linear Regression
def f(w, x, y):
    total = 0
    for i in range(len(x)):
        total += 1/2 * (np.dot(np.transpose(w), x[i]) - y[i])**2
    return total

def df(w, x, y):
    total = 0
    for i in range(len(x)):
        total += (np.dot(np.transpose(w), x[i]) - y[i]) * x[i]
    return total

auto = grad(f)

for i in range(10):
    w = np.random.random((2, 1))
    x = np.random.random((2, 2))
    y = np.random.random((2, 1))
    print(f'Input w: \n{w}')
    print(f'Input x: \n{x}')
    print(f'Input y: \n{y}')
    print(f'Auto: \n{auto(w, x, y)}\nTheoretical: \n{df(w, x, y)}\n')
    
```

```python
# Multivariate Gaussian Trace
def f(x, A):
    return np.exp(-np.trace(np.dot(np.dot(np.transpose(x), A), x)))

def df(x, A):
    return -np.exp(-np.trace(np.dot(np.dot(np.transpose(x), A), x))) * np.dot((A + np.transpose(A)), x)

auto = grad(f)

for i in range(10):
    x = np.random.random((2, 1))
    A = np.random.random((2, 2))
    print(f'Input x: \n{x}')
    print(f'Input A: \n{A}')
    print(f'Auto: \n{auto(x, A)}\nTheoretical: \n{df(x, A)}\n')
```

```python
# Multivariate Gaussian
def f(x, A):
    return np.exp(-np.dot(np.dot(np.transpose(x), A), x))

def df(x, A):
    return -np.exp(-np.dot(np.dot(np.transpose(x), A), x)) * np.dot((A + np.transpose(A)), x)

auto = grad(f)

for i in range(10):
    x = np.random.random((2, 1))
    A = np.random.random((2, 2))
    print(f'Input x: \n{x}')
    print(f'Input A: \n{A}')
    print(f'Auto: \n{auto(x, A)}\nTheoretical: \n{df(x, A)}\n')
```

```python
# Sigmoid
def f(x):
    return (1 + np.exp(-x))**-1

def df(x):
    return (1 + np.exp(-x))**-1 * (1 - (1 + np.exp(-x))**-1)

auto = grad(f)

for i in range(10):
    x = np.random.random()
    print(f'Input x: \n{x}')
    print(f'Auto: \n{auto(x)}\nTheoretical: \n{df(x)}\n')
```

```python
# Logistic Regression Objective
def f(w, x):
    return (1 + np.exp(-np.dot(np.transpose(w), x)))**-1

def df(w, x):
    z = np.dot(np.transpose(w), x)
    return (1 + np.exp(-z))**-1 * (1 - (1 + np.exp(-z))**-1) * x

auto = grad(f)

for i in range(10):
    w = np.random.random((2, 1))
    x = np.random.random((2, 1))
    print(f'Input w: \n{w}')
    print(f'Input x: \n{x}')
    print(f'Auto: \n{auto(w, x)}\nTheoretical: \n{df(w, x)}\n')
```

```python
# L1 Norm
def f(x):
    return sum(abs(i) for i in x)

def df(x):
    return np.sign(x)

auto = grad(f)

for i in range(10):
    x = np.random.randint(-10, 10, (2, 1))
    print(f'Input x: \n{x}')
    print(f'Auto: \n{auto(x)}\nTheoretical: \n{df(x)}\n')
```

```python
# L2 Norm
def f(x):
    return np.sqrt(sum(np.power(i, 2) for i in x))

def df(x):
    return x/np.sqrt(sum(np.power(i, 2) for i in x))

auto = grad(f)

for i in range(10):
    x = np.random.random((4, 1))
    print(f'Input x: \n{x}')
    print(f(x))
    print(f'Auto: \n{auto(x)}\nTheoretical: \n{df(x)}\n')
```

```python
# Gaussian MLE
def f(u, x, s=1):
    total = 0
    for i in range(len(x)):
        total += np.log(np.exp(-(x[i] - u)**2/(2*s**2))/(s * np.sqrt(2*np.pi)))
    return total

def df(u, x, s=1):
    total = 0
    for i in range(len(x)):
        total += x[i] - u
    return total / (s**2)

auto = grad(f)

for i in range(10):
    u = np.random.random()
    x = np.random.random((4, 1))
    print(f'Input u: \n{u}')
    print(f'Input x: \n{x}')
    print(f'Auto: \n{auto(u, x)}\nTheoretical: \n{df(u, x)}\n')
```

```python
# Exponential MLE
def f(l, x):
    total = 0
    for i in range(len(x)):
        total += np.log(l * np.exp(-l * x[i]))
    return total

def df(l, x):
    total = 0
    for i in range(len(x)):
        total += x[i]
    return len(x)/l - total

auto = grad(f)

for i in range(10):
    l = np.random.random()
    x = np.random.random((4, 1))
    print(f'Input u: \n{l}')
    print(f'Input x: \n{x}')
    print(f'Auto: \n{auto(l, x)}\nTheoretical: \n{df(l, x)}\n')
```

```python
# Bernoulli MLE
def f(p, a):
    total = 0
    for i in range(len(a)):
        total += np.log(np.power(p, a[i])) + np.log(np.power(1 - p, 1 - a[i]))
    return total

def df(p, a):
    total = 0
    for i in range(len(a)):
        total += a[i]/p - (1 - a[i])/(1 - p)
    return total

auto = grad(f)

for i in range(10):
    p = np.random.random()
    a = np.random.randint(0, 2, (2, 1))
    print(f'Input p: {p}')
    print(f'Input a: \n{a}')
    print(f'Auto: \n{auto(p, a)}\nTheoretical: \n{df(p, a)}\n')
```

```python
# Uniform MLE
def f(a, b, n):
    total = 0
    for _ in range(n):
        total += np.log(1/(b - a))
    return total

def df(a, b, n):
    return np.array([
        [n/(b-a)],
        [-n/(b-a)]
    ])

auto = grad(f)

for i in range(10):
    a = np.random.random()
    b = np.random.random()
    n = np.random.randint(1, 10)
    print(f'Input a: \n{a}')
    print(f'Input b: \n{b}')
    print(f'n = {n}')
    print(f'Auto: \n{auto(a, b, n)}\nTheoretical: \n{df(a, b, n)}\n')
```
