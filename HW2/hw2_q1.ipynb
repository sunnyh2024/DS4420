```python
# imports
import matplotlib.pyplot as plt 
import numpy as np
import utils
```

```python
# sample data
x = np.array([1, 2, 1.5, 3])

y = np.array([1, 1, 0, 2])
```

```python
# create the feature map and run gradient descent
X = utils.feature_map_quadratic(x)
print(f"Feature Map: {X}")
w = utils.gradient_descent(X, y, learning_rate=0.02, iters=1000)
print(f"w_quad: {w}")
```

```python
# plot the data points and the predicted function
x_plot = np.linspace(0, 4, 100)
y_plot = x_plot**2 * w[0] + x_plot * w[1] + w[2]

utils.plot(x, y, x_plot, y_plot, pred_label="quadratic model", scatter_label="data samples", plot_label="Quadratic Regression")
```
