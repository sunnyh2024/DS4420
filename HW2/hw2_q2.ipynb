```python
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt 
import utils
```

```python
# getting the data and the labels
x = genfromtxt("time_with_loved_ones.csv")
y = genfromtxt("happiness.csv")

print(x.shape)
print(y.shape)
```

```python
# transform the data to feature map and run gradient descent
X = utils.feature_map_linear(x)
print(X.shape)
w = utils.gradient_descent(X, y, learning_rate=0.002, iters=20000)
print(f"output: {w}")
```

```python
# plot the data points with the predicted function
x_plot = np.linspace(0, 20, 3000)
y_plot = x_plot * w[0] + w[1]

utils.plot(x, y, x_plot, y_plot, pred_label="linear model", scatter_label="data samples", plot_label="Linear Regression")
```

```python
# get the happiness for 15 hr/wk
print(f"weights: {w}")
print(f"Prediction: {15*w[0] + w[1]}")
```
