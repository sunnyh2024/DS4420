```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
sat = pd.read_csv("SAT.csv")
lunch = pd.read_csv("lunch_wait_time.csv")
student_age = pd.read_csv("student_age.csv")

data = [(sat, "SAT"), (lunch, "lunch wait times"), (student_age, "student age")]
```

```python
def activity(df, title):
    plt.hist(df, 20)
    plt.title(title)
    plt.show()

    print(sum(df.to_numpy())/len(df))
```

```python
for sample in data:
    df, title = sample
    activity(df, title)
```

```python
# probability of sat > 1300
sat_numpy = sat.to_numpy()

prob = sum([1 for sample in sat_numpy if float(sample) > 1300])/len(sat_numpy)
prob
```

```python
# probability of wait time > 3 min
lunch_numpy = lunch.to_numpy()

prob = sum([1 for sample in lunch_numpy if float(sample) > 3])/len(lunch_numpy)
prob
```

```python
# probability of age < 21
student_age_numpy = student_age.to_numpy()

prob = sum([1 for sample in student_age_numpy if float(sample) < 21])/len(student_age_numpy)
prob
```
